# Import Data from a URL

```python
import requests
from io import StringIO

url = "https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv"
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"} # You need to add this line while importing data from any online resource otherwise it'll consider you as a bot
req = requests.get(url, headers=headers)
data = StringIO(req.text)

pd.read_csv(data)
```

## Handle Dates

```python
pd.read_csv('IPL Matches 2008-2020.csv', parse_dates=['date']).info()
```

Look above the dtype of "Date" column has been changes to the Datetime. That can be used to fulfill various purpose that wouldn't been possible while using int.

```python
def rename(name):
    if name == "Mumbai Indians":
        return "MI"
    elif name == "Royal Challengers Bangalore":
        return "RCB"
    else:
        return name #you can write functions like this that'll be reusable throughtout the code
```

```python
rename("Mumbai Indians")
```

## Convertors

Converters are functions that we can apply on any column to manipulate the data. Like we created this rename function.

```python
pd.read_csv('IPL Matches 2008-2020.csv',converters={'team1':rename})
```

## na_values

```python
pd.read_csv('titanic.csv',na_values=['female'])
```

na_values parameter converts the passed value into a NaN value

## Loading a huge dataset in chunks

```python
large_df = pd.read_csv('zomato.csv',chunksize=1000, encoding='latin-1')
```

```python
for chunk in large_df:
    print(chunk.shape)
```

As the dataset becomes huge in size. Our machine won't be able to work. So, We can divide our data into smaller chunks and process it
