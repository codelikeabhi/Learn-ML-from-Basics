  # <span style="color:red"> Let's Understand Bivariate Analysis</span>

<span style='color:blue'> As the name suggests in this type of analysis we can perform analysis on two columns of data at a time.According to their data types. In order to analysis the data in the better manner we need to visualize the data. 
Some plotting created mainly for Bivariate Analysis are:
- Scatter Plot
- Bar Plot
- Box Plot
- Dist Plot
- Heatmap
- Clustermap
- PairPlot, etc.</span>

```python
import pandas as pd
import seaborn as sns
```

### Let's import some dataset from Seaborn library

```python
tips = sns.load_dataset('tips')
```

```python
tips.sample(7)
```

```python
flights = sns.load_dataset('flights')
```

```python
flights.sample(9)
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.sample(7)
```

```python
titanic = pd.read_csv('titanic.csv')
```

```python
titanic.sample(8)
```

# 1. Scatter Plot

<span style='color:blue'>Scatter plot represents the relationship between two variables in a data-set (Numerical -Numerical)</span>

```python
sns.scatterplot(data=tips, x='total_bill', y='tip')
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='time')
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day', style='time')
```

```python
sns.relplot(data=tips, x='total_bill', y='tip', hue='day', style='day', col='time', kind='scatter')
```

# 2. Bar Plot

<span style='color:blue'>Bar plot represents the category of data with rectangular bars with lengths and heights proportional to value they're representing</span>

```python
sns.barplot(titanic, x='Age', y='Embarked')
```

```python
sns.barplot(titanic, x='Age', y='Embarked', hue='Sex', legend=False)
```

```python
sns.barplot(titanic, x='Age', y='Embarked', hue='Sex')
```

```python
flight_w = flights.pivot(index ='year', columns='month', values='passengers')
sns.barplot(flight_w)
```

# 3. Box Plot

<span style='color:blue'>Box plot is used to display the distribution of the data and the outliers present in the data</span>

```python
sns.boxplot(x=titanic['Age'])
```

```python
sns.boxplot(x=titanic['Age'], y=titanic['Sex'])
```

```python
sns.boxplot(x=titanic['Age'], y=titanic['Sex'], hue=titanic['Survived'])
```

```python
sns.boxplot(x=titanic['Age'], y=titanic['Sex'], hue=titanic['Survived'], fill=False, gap=0.1)
```

```python
sns.boxplot(x=titanic['Sex'], y=titanic['Age'], whis=(0, 100), width=0.5)
```

```python
sns.boxplot(x=titanic['Age'], y=titanic['Sex'], hue=titanic['Survived'], fill=False, gap=0.1, notch=True)
```

# 4. Displot (Updates version of Distplot)

<span style='color:blue'>Displot earlier called as DistPlot is used to depict the variation in data distribution</span>

```python
sns.displot(data = iris, x='petal_length')
```

```python
sns.displot(data = iris, x='petal_length', kind='kde')
```

```python
sns.displot(data = iris, x='petal_width', kind='ecdf')
```

```python
sns.displot(data = iris, x='petal_length', kde=True)
```

```python
sns.displot(data = iris, x='petal_length', y='petal_width')
```

```python
sns.displot(data = iris, x='petal_length', y='petal_width', kind='kde', rug=True)
```

```python
sns.displot(data = iris, x='petal_length', y='petal_width', hue='species', kind='kde')
```

```python
sns.displot(data = iris, x='petal_length', y='petal_width', col='species', kind='kde')
```

# 5. Heat Map

<span style='color:blue'>HeatMap uses graphical representation of data that uses a system of color coding to represent different values present in data</span>

```python
sns.heatmap(flight_w, cmap='crest')
```

```python
sns.heatmap(flight_w, annot=True, fmt='.1f')
```

```python
sns.heatmap(flight_w, annot=flight_w.rank(axis='columns'), linewidth=0.5)
```

```python
sns.heatmap(flight_w, annot=flight_w.rank(axis='columns'), linewidth=0.5, vmin=200, vmax=500)
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

# 6. ClusterMap

<span style='color:blue'>Cluster Map is used to group data based on similarity</span>

```python
sns.clustermap(iris.iloc[0:, :4])
```

```python
sns.clustermap(iris.iloc[0:, :4], figsize=(7,5), row_cluster=False, cbar_pos=(0, .2, .03, .5))
```

```python
sns.clustermap(iris.iloc[0:, :4], standard_scale=1)
```

```python
sns.clustermap(flight_w)
```

# 7. Pair Plot

```python
<span style='color:blue'>Pair Plot is used to plot pairwise relationship present in the data</span>
```

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris,  hue='species')
```

```python
sns.pairplot(iris, kind='kde', height=1.5)
```

```python
g = sns.pairplot(iris, diag_kind="kde")
g.map_lower(sns.kdeplot, levels=4, color=".2")
```

# 8. Line Plot

<span style='color:blue'>Line plot is used to display data as points or check marks above a number line, showing the frequency of each value</span>

```python
sns.lineplot(data=flights, x='year', y='passengers')
```

```python
sns.lineplot(flight_w)
```

```python
fmri = sns.load_dataset('fmri')
```

```python
fmri.sample(8)
```

```python
sns.lineplot(data=fmri, x='timepoint', y='signal', hue='region', style='event')
```
